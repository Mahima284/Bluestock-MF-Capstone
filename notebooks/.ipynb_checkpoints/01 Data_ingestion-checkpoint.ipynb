```python
response=requests.get("https://api.mfapi.in/mf/125497")###API calll
data= response.json()## response is in the form of json this converts into python
print("status code :",response.status_code)## check the status code
##json has 2 parts meta-scheme name.fund house,category,PLAN TYPE
print("Keys in response :",data.keys()) ##data contains a list daily NAV values
print("First 3 NAV records :",data['data'][:3])
```

### Task 4 Fetch live NAV from mfapi.in
-called mfapi.in API for sheme code 125497
-API returned JSON with 3 keys meta(scheme details),data(list of daily NAV values), and status(tells if request is succesfull)
-status code 200 shows successfull connection
Total NAV records fetch =3093 rows

```python
nav_records=data['data']## data['data'] is a list of dictionaries-each dictionar has a date and nav value
df_hdfc=pd.DataFrame(nav_records)## converts the list of dictionary to data frame
df_hdfc['amfi_code']=125497
df_hdfc['scheme_name']=data['meta']['scheme_name']##adds two new columns amficode and scheme name
print(df_hdfc.shape)
print(df_hdfc.head())
```

```python
df_hdfc.to_csv(os.path.join(RAW_DIR,'live_nav_125497.csv'),index=False)
print("Saved successfully")
```

saved as live_nav_125497.csv in data raw
